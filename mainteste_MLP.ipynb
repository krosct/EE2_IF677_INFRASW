# DataLoader
O data Loader será utilizado para poder pegar os dados aleatoriamente e controlar o "Batch Size"

```python
class MNIST(Dataset):

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __len__(self):
        return len(self.y)

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
```

```python
train_dl = DataLoader(MNIST(train_X, train_y), batch_size=256, shuffle = True)
test_dl = DataLoader(MNIST(test_X, test_y), batch_size= 64)
```

# MLP

## Criando o Modelo de MLP

```python
class MLP_Model(nn.Module):
    def __init__(self, epoch, dropout1, cat, saidaL1 = 0, saidaL2 = 0, saidaL3 = 0):
        super(MLP_Model, self).__init__()

        self.A1 = nn.Flatten()
        self.A2 = nn.ReLU()
        self.D1 = nn.Dropout(dropout1)
        self.L1 = nn.Linear(28*28, saidaL1)
        self.L2 = nn.Linear(saidaL1, saidaL2)
        self.L3 = nn.Linear(saidaL2, saidaL3)
        self.cat = cat

    def forward(self, x):
      if self.cat == 3:
        x = self.A1(x)
        x = self.L1(x)
        x = self.A2(x)
        x = self.L2(x)
        x = self.A2(x)
        x = self.L3(x)
      elif self.cat == 2:
        x = self.A1(x)
        x = self.L1(x)
        x = self.A2(x)
        x = self.L2(x)
      elif self.cat == 1:
        x = self.A1(x)
        x = self.L1(x)

        return x
```

## Treinando o modelo


mlp:
nr de camadas ocultas 1 2 3
neuronios por camada 20 40 80
função de custo MSE cross entropy
função de ativação ReLU Sigmoid tnh
dropout 0 0.1 0.3 0.5
learning rate 3-5 3-2 3-10
epocas 20 50 80 100
regularização dropout sem
cnn:
tamanho kernel 2 4 8
stride kernel 1 2
função de ativação Relu sigmoid tnh
função de custo cross entropy MSE
nr camadas convolução 1 2
tamanho kernel max polling 2 4
nr mapa caracteristica 5 10 15


```python
def training(N_Epochs, model, loss_fn, opt):

    loss_list = []

    for epoch in tqdm(range(N_Epochs+1)):
        for xb, yb in train_dl:

            y_pred = model(xb.float())
            loss   = loss_fn(y_pred, yb.long())

            opt.zero_grad()
            loss.backward()
            opt.step()

        loss_list.append(loss.item())

    # plt.figure(figsize = (3,2))
    # plt.title("cost decay")
    # plt.plot(loss_list)
    # plt.xlabel("epoch")
    # plt.ylabel("cost")

```

```python
camadas = [1, 2, 3]
epoch1 = [20, 50, 80, 100]
dropout1 = [0, 0.1, 0.3, 0.5]
saidas = [20, 40, 80]
custo = [nn.CrossEntropyLoss(), nn.MSELoss()]
ativacao = [nn.ReLU(), nn.Sigmoid(), nn.Tanh()]
learning_rate = [0.03, 0.00003, 0.0000000003]

#mlp:
#nr de camadas ocultas 1 2 3
#neuronios por camada 20 40 80
#função de custo MSE cross entropy
#função de ativação ReLU Sigmoid tnh
#dropout 0 0.1 0.3 0.5
#learning rate 3-5 3-2 3-10
#epocas 20 50 80 100
#regularização dropout sem
#cnn:
#tamanho kernel 2 4 8
#stride kernel 1 2
#função de ativação Relu sigmoid tnh
#função de custo cross entropy MSE
#nr camadas convolução 1 2
#tamanho kernel max polling 2 4
#nr mapa caracteristica 5 10 15

result_list = []

for cam in camadas:
  for a in epoch1:
    for b in dropout1:
      for d in custo:
        for e in ativacao:
          for f in learning_rate:
            model = MLP_Model(a, b, cam, saidas[2], saidas[1], saidas[0])
            loss = d
            opt = torch.optim.Adam(model.parameters(), lr = f)
            training(a, model, loss, opt)

            with torch.no_grad():
              model.eval()
              y_pred = []
              y_true = []

              for xb, yb in test_dl:
                  y_predb = model(xb.float())
                  y_pred.append(y_predb)
                  y_true.append(yb)

              y_pred = torch.cat(y_pred)
              y_true = torch.cat(y_true)
              yf = torch.argmax(y_pred, dim=1)

              result_list.append(classification_report(y_true, yf))

print(result_list)
```

## Avaliando a efetividade do Modelo

```python
with torch.no_grad():

    model.eval()

    y_pred = []
    y_true = []

    for xb, yb in test_dl:
        y_predb = model(xb.float())
        y_pred.append(y_predb)
        y_true.append(yb)

    y_pred = torch.cat(y_pred)
    y_true = torch.cat(y_true)

    yf     = torch.argmax(y_pred, dim=1)

    print(classification_report(y_true, yf))
```

# CNN

## Alteração no formato dos dados para utilizar a CNN

```python

train_X = train_X[:, None, :, :]  # Adiciona o canal (batch_size, 1, 28, 28)
test_X = test_X[:, None, :, :]

train_dl = DataLoader(MNIST(train_X, train_y), batch_size=256, shuffle = True)
test_dl = DataLoader(MNIST(test_X, test_y), batch_size= 64)
```

## Criando o Modelo CNN

```python


# Criar uma classe que herda o modulo nn.Module

class CNN_Model(nn.Module):

    def __init__(self):
        super().__init__()

        # transformar a entrada em um único tensor unidimensional

        self.Fl = nn.Flatten()

        # a primeira camada de convolução
        # o parâmetro in_channels é para identificar as dimensões que a representação da sua imagem possuí, ex RGB = 3, gray_scale = 1, ou os mapas de características dado por camadas anteriores
        # e o out_channels é para a saída gerada pela unidade computacional, maiores valores gerarão mais mapas de características
        # o kernel e o stride podem assumir valores diferentes, mas por fins de simplificação iremos apenas passar inteiros

        self.C1 = nn.Conv2d(in_channels = 1, out_channels = 5, kernel_size = 4, stride = 1 )

        # após a operação de convolução, virá a operação de pooling
        # existem diferentes formas pela biblioteca para fazer a operação de pooling, mas utilizaremos o Max

        self.P1 = nn.MaxPool2d(kernel_size = 2)

        #segunda camada de convolução + pooling

        self.C2 = nn.Conv2d(in_channels = 5, out_channels = 10, kernel_size = 4, stride = 1)
        self.P2 = nn.MaxPool2d(kernel_size = 4)


        # Após n operações de convolução e pooling a imagem será achatada pela operação Flatten
        # e será usado em uma rede MLP
        # o pytorch não oferece maneiras dinâmicas de fazer alteração nos parâmetros da rede após a criação
        # então o valor das entradas e saídas após cada camada devem ser calculadas: output_size=((input_size−kernel_size+2×padding)/stride) + 1
        # para a primeira camada após a convolução temos output_size = 25
        # para a primeira camada após o pooling temos output_size =  22
        # para a segunda camada após a convolução temos output_size = 19
        # para a segunda camada após o pooling temos output_size = 16
        # então para a rede MLP teremos uma quantidade de entradas 16x16

        #primeira camada terá 100 unidades computacionais
        self.L1 = nn.Linear(10*2*2, 100)

        #segunda camada para 50 unidades computacionais
        self.L2 = nn.Linear(100, 50)

        #terceira camada para 10
        self.L3 = nn.Linear(50, 10)


    def forward(self, x):


        x = torch.relu(self.C1(x))  # Primeira convolução + ReLU
        x = self.P1(x)              # Primeiro pooling
        x = torch.relu(self.C2(x))  # Segunda convolução + ReLU
        x = self.P2(x)              # Segundo pooling
        x = self.Fl(x)              # Flatten
        x = torch.relu(self.L1(x))  # Primeira camada MLP
        x = torch.relu(self.L2(x))  # Segunda camada MLP
        x = self.L3(x)              # Saída final

        return x
```

## Treinando o Modelo CNN



```python
epoch = 50
model =  CNN_Model()
loss = nn.CrossEntropyLoss()
opt = torch.optim.Adam( model.parameters(), lr=0.00003 )
training(epoch,model, loss, opt)
```

## Avaliando a CNN

```python
with torch.no_grad():

    model.eval()

    y_pred = []
    y_true = []

    for xb, yb in test_dl:
        y_predb = model(xb.float())
        y_pred.append(y_predb)
        y_true.append(yb)

    y_pred = torch.cat(y_pred)
    y_true = torch.cat(y_true)

    yf     = torch.argmax(y_pred, dim=1)

    print(classification_report(y_true, yf))
```
